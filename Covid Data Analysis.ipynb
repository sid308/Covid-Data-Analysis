```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
dataset = pd.read_csv('/content/covid19_Confirmed_dataset.csv')
dataset.head()
dataset.shape
#Delete the useless columns
df=dataset.drop(columns=["Lat", "Long"], inplace=True)
```

```python
#Aggregate The rows by country
corona_dataset_aggregated=dataset.groupby("Country/Region").sum()
corona_dataset_aggregated.shape
```

```python
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Andorra"].plot()
plt.legend()
```

```python
#Calculate a good measure
corona_dataset_aggregated.loc["Spain"][:3].plot()
```

```python
#Calculate the first derivative of the curve
corona_dataset_aggregated.loc["India"].diff().plot()
```

```python
#maximum infection rate
corona_dataset_aggregated.loc["India"].diff().max()
```
